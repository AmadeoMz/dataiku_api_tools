```python
# %pylab inline
import dataiku
import dataikuapi

from dataiku import pandasutils as pdu
import pandas as pd

import time
```

```python
client = dataiku.api_client() # Connect to dataiku
project = client.get_project('project_name') #Take project
all_scenarios = pd.DataFrame(project.list_scenarios()) #Take scenaries 

# Tags filtering function 
def tag_filter(tag_name):
    """Take the tag and create a mask that filters the dataframe.
    Args
    tag_name (str): name of scenario tag
    
    Returns
    all_tags_mask (list): Boolean list with True when tag_name is in the scenario's tags list or False otherwise.
    """
    all_tags_mask = list()
    for i in range(len(all_scenarios)): 
        all_tags_mask.append(tag_name in all_scenarios.loc[i,'tags'])
    return all_tags_mask

#### Filtramos por la tag 100Tablas_R1
tag_mask = tag_filter('100Tablas_R1') 
filtered_scenarios_ids = all_scenarios[tag_mask]['id'] # Ids from scenarios in which we are interested

```

```python
alphabetical = {'Name':[], 'Duration':[]}
######### Obtenemos detalles de la última run como el resultado de la misma y los tiempos para calcular la duración
for scenario_id in filtered_scenarios_ids:
    scenario = project.get_scenario(scenario_id)
    
    ##### Run info
    last_run = scenario.get_last_finished_run()
    outcome = last_run.outcome # run's result
    
    ##### Time
    start = last_run.start_time
    end = last_run.end_time
    duration = (str(end - start))
    
    ##### Action
    if outcome == 'SUCCESS':
        if duration[0] == '0':
            alphabetical['Duration'].append('{}m {}s'.format(duration[2:4],duration[-2:-1]))
        else:
            alphabetical['Duration'].append('{}h {}m {}s'.format(duration[0],duration[2:4],duration[-2:-1]))
    else:
        alphabetical['Duration'].append('fallo')
        
    #### name appending creation
    name = last_run.get_details()['stepRuns'][0]['scenarioRun']['scenario']['params']['steps'][0]['params']['config']['parameters']['table-name']
    nombre = str(name).lower()
    alphabetical['Name'].append(nombre)
        

alphabetical_df = pd.DataFrame(alphabetical)
print(alphabetical_df.sort_values('Name')['Duration'].to_string(index=False))
```
